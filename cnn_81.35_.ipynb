```python
# pip install tensorflow scikit-learn

from google.colab import drive
drive.mount('/content/drive')

# 0. SETUP
import os, shutil, random, numpy as np, matplotlib.pyplot as plt, tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv2D, MaxPooling2D, GlobalAveragePooling2D,
                                     Dense, Dropout, BatchNormalization)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras import mixed_precision, regularizers
from sklearn.metrics import classification_report, confusion_matrix, f1_score, recall_score, precision_score, accuracy_score

SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)
mixed_precision.set_global_policy('mixed_float16')

# ==========================
# 1ï¸âƒ£ DOWNLOAD Food-101
# ==========================
# XÃ³a thÆ° má»¥c cÅ© náº¿u cÃ³
if os.path.exists("food-101"):
    print("ğŸ—‘ï¸ Removing old food-101 folder...")
    shutil.rmtree("food-101")
if os.path.exists("food-101-subset"):
    print("ğŸ—‘ï¸ Removing old food-101-subset folder...")
    shutil.rmtree("food-101-subset")
if os.path.exists("food-101-split"):
    print("ğŸ—‘ï¸ Removing old food-101-split folder...")
    shutil.rmtree("food-101-split")

# XÃ³a file tar.gz cÅ© náº¿u bá»‹ lá»—i vÃ  táº£i láº¡i
if os.path.exists("food-101.tar.gz"):
    print("ğŸ—‘ï¸ Removing corrupted food-101.tar.gz...")
    os.remove("food-101.tar.gz")

# Download dataset
print("ğŸ“¥ Downloading Food-101 dataset (~5GB)...")
!wget --show-progress http://data.vision.ee.ethz.ch/cvl/food-101.tar.gz

# Verify file size (should be ~5GB)
import os
file_size = os.path.getsize("food-101.tar.gz") / (1024**3)
print(f"âœ… Downloaded file size: {file_size:.2f} GB")

if file_size < 4.5:
    raise Exception("âŒ Download incomplete! File too small. Please run again.")

# Extract dataset
print("ğŸ“¦ Extracting Food-101 dataset...")
!tar -xzf food-101.tar.gz

# Verify extraction
if not os.path.exists("food-101/images"):
    raise Exception("âŒ Extraction failed! food-101/images not found.")

print("âœ… Dataset extracted successfully!")

data_dir = "food-101/images"
subset_dir = "food-101-subset"
split_dir = "food-101-split"
os.makedirs(subset_dir, exist_ok=True)

# ==========================
# 2ï¸âƒ£ Select 20 classes + copy images
# ==========================
selected_classes = [
    'guacamole', 'hot_and_sour_soup', 'ravioli', 'caprese_salad', 'chocolate_mousse',
    'strawberry_shortcake', 'oysters', 'sashimi', 'poutine', 'deviled_eggs',
    'huevos_rancheros', 'prime_rib', 'panna_cotta', 'tacos', 'beef_carpaccio',
    'donuts', 'waffles', 'bread_pudding', 'beignets', 'mussels'
]

for cls in selected_classes:
    src_folder = os.path.join(data_dir, cls)
    dst_folder = os.path.join(subset_dir, cls)
    os.makedirs(dst_folder, exist_ok=True)
    imgs = os.listdir(src_folder)
    for img in imgs:
        shutil.copy(os.path.join(src_folder, img), os.path.join(dst_folder, img))

print("âœ… Copied all images for", len(selected_classes), "classes")

# ==========================
# 3ï¸âƒ£ SPLIT 80/10/10 (train/val/test)
# ==========================
os.makedirs(split_dir, exist_ok=True)
for split in ['train', 'val', 'test']:
    os.makedirs(os.path.join(split_dir, split), exist_ok=True)

for cls in selected_classes:
    src_folder = os.path.join(subset_dir, cls)
    imgs = os.listdir(src_folder)
    random.shuffle(imgs)

    n = len(imgs)
    train_split = int(0.8 * n)
    val_split = int(0.9 * n)

    for i, img in enumerate(imgs):
        if i < train_split:
            folder = 'train'
        elif i < val_split:
            folder = 'val'
        else:
            folder = 'test'

        dst_folder = os.path.join(split_dir, folder, cls)
        os.makedirs(dst_folder, exist_ok=True)
        shutil.copy(os.path.join(src_folder, img), os.path.join(dst_folder, img))

print("âœ… Dataset split into train/val/test with 80/10/10 ratio")

for split in ['train', 'val', 'test']:
    count = sum([len(os.listdir(os.path.join(split_dir, split, cls)))
                 for cls in selected_classes])
    print(f"   {split}: {count} images")

# ==========================
# 4ï¸âƒ£ DATA PIPELINE (NO AUGMENTATION)
# ==========================
IMG_SIZE = 224  # TÄƒng lÃªn Ä‘á»ƒ model há»c tá»‘t hÆ¡n
BATCH_SIZE = 64  # Giáº£m Ä‘á»ƒ trÃ¡nh overfitting vÃ  cáº£i thiá»‡n gradient
NUM_CLASSES = len(selected_classes)

# Chá»‰ rescale, khÃ´ng augmentation
datagen = ImageDataGenerator(rescale=1./255)

train_gen = datagen.flow_from_directory(
    os.path.join(split_dir, 'train'),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True,
    seed=SEED
)

val_gen = datagen.flow_from_directory(
    os.path.join(split_dir, 'val'),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    seed=SEED
)

test_gen = datagen.flow_from_directory(
    os.path.join(split_dir, 'test'),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    seed=SEED
)

print(f"âœ… Data generators ready:")
print(f"   Train: {train_gen.samples} samples")
print(f"   Val: {val_gen.samples} samples")
print(f"   Test: {test_gen.samples} samples")

# ==========================
# 5ï¸âƒ£ IMPROVED CNN - Tá»‘i Æ°u cho khÃ´ng overfitting
# ==========================
# Sá»­ dá»¥ng L2 regularization máº¡nh
l2_reg = regularizers.l2(1e-3)

model = Sequential()

# BLOCK 1 - 32 filters (giáº£m tá»« 64)
model.add(Conv2D(32, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg,
                 input_shape=(IMG_SIZE, IMG_SIZE, 3)))
model.add(BatchNormalization())
model.add(Conv2D(32, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.3))

# BLOCK 2 - 64 filters
model.add(Conv2D(64, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(Conv2D(64, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.35))

# BLOCK 3 - 128 filters
model.add(Conv2D(128, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(Conv2D(128, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.4))

# BLOCK 4 - 256 filters (giáº£m tá»« 512)
model.add(Conv2D(256, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(Conv2D(256, (3,3), padding='same', activation='relu',
                 kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.45))

# GLOBAL POOLING â†’ DENSE
model.add(GlobalAveragePooling2D())
model.add(Dense(256, activation='relu', kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu', kernel_regularizer=l2_reg))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(NUM_CLASSES, activation='softmax', dtype='float32'))

model.summary()

# ==========================
# 6ï¸âƒ£ COMPILE vá»›i Label Smoothing
# ==========================
# Label smoothing giÃºp trÃ¡nh overfitting
loss_fn = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.15)
optimizer = AdamW(learning_rate=1e-3, weight_decay=1e-4)
model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

# ==========================
# 7ï¸âƒ£ CALLBACKS
# ==========================
# Early stopping dá»±a vÃ o val_loss Ä‘á»ƒ trÃ¡nh overfitting
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=25,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=8,
    min_lr=1e-7,
    verbose=1
)

# Cosine Decay Learning Rate
total_epochs = 100
def cosine_decay(epoch):
    initial_lr = 1e-3
    min_lr = 1e-6
    decay_steps = total_epochs
    alpha = min_lr / initial_lr
    cosine_decay = 0.5 * (1 + np.cos(np.pi * epoch / decay_steps))
    decayed = (1 - alpha) * cosine_decay + alpha
    return initial_lr * decayed

lr_sched = tf.keras.callbacks.LearningRateScheduler(cosine_decay, verbose=0)

# ==========================
# 8ï¸âƒ£ TRAIN
# ==========================
print("\nğŸš€ Starting training...")
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=total_epochs,
    callbacks=[early_stop, reduce_lr, lr_sched],
    verbose=1
)

# ==========================
# 9ï¸âƒ£ EVALUATE & FULL METRICS
# ==========================
print("\n" + "="*60)
print("ğŸ“Š EVALUATING ON TEST SET")
print("="*60)

test_loss, test_acc = model.evaluate(test_gen, verbose=0)
print(f"\nâœ… Test Loss: {test_loss:.4f}")
print(f"âœ… Test Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)")

# Get predictions
test_gen.reset()
y_true = test_gen.classes
y_pred_prob = model.predict(test_gen, verbose=0)
y_pred = np.argmax(y_pred_prob, axis=1)

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision_macro = precision_score(y_true, y_pred, average='macro', zero_division=0)
recall_macro = recall_score(y_true, y_pred, average='macro', zero_division=0)
f1_macro = f1_score(y_true, y_pred, average='macro', zero_division=0)

precision_weighted = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall_weighted = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1_weighted = f1_score(y_true, y_pred, average='weighted', zero_division=0)

print("\n" + "="*60)
print("ğŸ“ˆ TEST SET METRICS")
print("="*60)
print(f"Accuracy:           {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"\nMacro Average:")
print(f"  Precision:        {precision_macro:.4f}")
print(f"  Recall:           {recall_macro:.4f}")
print(f"  F1-Score:         {f1_macro:.4f}")
print(f"\nWeighted Average:")
print(f"  Precision:        {precision_weighted:.4f}")
print(f"  Recall:           {recall_weighted:.4f}")
print(f"  F1-Score:         {f1_weighted:.4f}")

# Classification Report
print("\n" + "="*60)
print("ğŸ“‹ CLASSIFICATION REPORT")
print("="*60)
print(classification_report(y_true, y_pred, target_names=selected_classes, digits=4))

# Confusion Matrix
print("\n" + "="*60)
print("ğŸŸ¢ CONFUSION MATRIX")
print("="*60)
cm = confusion_matrix(y_true, y_pred)
print(cm)

# Plot confusion matrix
plt.figure(figsize=(14, 12))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix', fontsize=16)
plt.colorbar()
tick_marks = np.arange(len(selected_classes))
plt.xticks(tick_marks, selected_classes, rotation=90)
plt.yticks(tick_marks, selected_classes)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()

# Plot training history
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Val Loss', linewidth=2)
plt.title('Loss History', fontsize=14)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2)
plt.title('Accuracy History', fontsize=14)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ==========================
# ğŸ”Ÿ SAVE MODEL
# ==========================
save_path = '/content/drive/MyDrive/cnn4_food20_optimized_noaug.h5'
model.save(save_path)
print(f"\nâœ… Saved model to {save_path}")

# Save metrics to text file
metrics_path = '/content/drive/MyDrive/cnn4_food20_optimized_metrics.txt'
with open(metrics_path, 'w') as f:
    f.write("="*60 + "\n")
    f.write("TEST SET METRICS (Optimized CNN - No Augmentation)\n")
    f.write("="*60 + "\n")
    f.write(f"Accuracy:           {accuracy:.4f} ({accuracy*100:.2f}%)\n")
    f.write(f"\nMacro Average:\n")
    f.write(f"  Precision:        {precision_macro:.4f}\n")
    f.write(f"  Recall:           {recall_macro:.4f}\n")
    f.write(f"  F1-Score:         {f1_macro:.4f}\n")
    f.write(f"\nWeighted Average:\n")
    f.write(f"  Precision:        {precision_weighted:.4f}\n")
    f.write(f"  Recall:           {recall_weighted:.4f}\n")
    f.write(f"  F1-Score:         {f1_weighted:.4f}\n")
    f.write("\n" + "="*60 + "\n")
    f.write("CLASSIFICATION REPORT\n")
    f.write("="*60 + "\n")
    f.write(classification_report(y_true, y_pred, target_names=selected_classes, digits=4))

print(f"âœ… Saved metrics to {metrics_path}")

print("\n" + "="*60)
print("âœ… TRAINING AND EVALUATION COMPLETED!")
print("="*60)
```
